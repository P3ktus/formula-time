```python
races_df = pd.read_csv("./races.csv")
```

```python
def toDatetime (date, time):
    if(time!="\\N"):
        d = datetime.strptime(date + " " + time, '%d/%m/%y %H:%M:%S')
    else:
        d = datetime.strptime(date, '%d/%m/%y')

    if d > datetime.strptime("1/1/2040", '%d/%m/%Y'):
        d = datetime(d.year-100, d.month, d.day, d.hour, d.minute, d.second, d.microsecond, d.tzinfo)
    
    return d.isoformat()+'Z'
```

```python
races_df['datetime'] = races_df.apply (lambda row: toDatetime(row["date"], row["time"]), axis=1)
```

```python
races_df.to_sql("races", engine, if_exists="append", index=False)
```
